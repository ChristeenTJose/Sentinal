# Import required libraraies

```
import cv2
import matplotlib.pyplot as plt
```

# Loading an image

```
#0 (or) IMREAD_GRAYSCALE
Image = cv2.imread("face.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure();
plt.imshow(Image)
#-1 (or) IMREAD_UNCHANGED
Image = cv2.imread("face.jpg", cv2.IMREAD_UNCHANGED)
plt.figure();
plt.imshow(Image)
plt.show()
```

```
#1 (or) IMREAD_COLOR
Image = cv2.imread("face.jpg", cv2.IMREAD_COLOR)
plt.imshow(Image)
plt.show()
```

# Convert the image to Grayscale

```
#Convert to grayscale
Image = cv2.cvtColor(Image,cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(Image)
#Same image 
cv2.imwrite('result1.jpg',Image)
Image=cv2.imread('result1.jpg')
plt.figure()
plt.imshow(Image)
plt.show()
print("Note:\nThis is because openCV uses BGR instead of RGB")
```

# Load haarcascade_frontalface_default xml file for detecting faces

```
model = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

#Detecting faces in the image 

```
faces = model.detectMultiScale(Image, 1.1, 5)
print("Number of faces found: ",len(faces))
```

#Visualising the detected faces 





##Drawing boundaries


```
#Drawing boundaries
for (x,y,W,H) in faces:
    cv2.rectangle(Image,(x,y),(x+W,y+H),(255,0,0),2)
```

##Saving 

```
#Saving
cv2.imwrite('result2.jpg',Image)
```

##Visualising 

```
Image=cv2.imread('result2.jpg')
plt.imshow(Image)
plt.show()
```
